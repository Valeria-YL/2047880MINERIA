```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("sumedh1507/blood-donor-dataset")

print("Path to dataset files:", path)
```

```python
import os
os.listdir(path)
```

Convertimos el csv a dataframe y mostramos las primeras 5 filas

```python
import pandas as pd

df = pd.read_csv(os.path.join(path, "blood_donor_dataset.csv"))
df.head()
```

Nos aseguramos que el dataset este limpio

```python
print('filas duplicadas',df.duplicated().sum())
print(df.isnull().sum())
```

Estadísticas Descriptivas

```python
print("Estadísticas Descriptivas Valores Númericos:")
print(df.describe())
print("\n")
print("Estadísticas Descriptivas Valores Categóricos:")
print(df.describe(include='object'))
```

Estadísticas Agrupadas

```python
print("Total de donantes por grupo sanguíneo:")

df.groupby('blood_group').agg({
    'donor_id': 'count',
    'number_of_donation': 'sum',
    'pints_donated': 'sum'
})
```

```python
print("Distribución de donantes por ciudad:")
df.groupby('city').agg({'donor_id': 'count', 'pints_donated': 'sum'})
```

```python
print("Disponibilidad según los tipos de sangre:")
df.groupby(['blood_group', 'availability']).agg({
    'donor_id': 'count',
    'pints_donated': 'sum'
}).unstack()
```

```python
print("Tendencias de Donantes y Donaciones por Año:")
df['created_at'] = pd.to_datetime(df['created_at'])
df.groupby(df['created_at'].dt.year).agg({
    'donor_id': 'count',
    'pints_donated': 'sum'
})
```
